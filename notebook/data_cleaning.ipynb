### Drop Duplicates

```python
df = df.drop_duplicates()
df.shape
```

### Team & Contract cleaning üßº

```python
# This is a problem area, we need to make two different columns out of this one column 
df['Team & Contract'].head()
```

```python
# Extract "Team" and "Contract" using regular expression.
df[['Team','Contract']] = df['Team & Contract'].str.extract(r'\n\n\n\n(.+)\n(.+)\n\n')
```

```python
# Check to see how it looks 
df[['Team','Contract']].head()
```

```python
# drop the useless column
df.drop(columns = ['Team & Contract'], inplace=True)
```

#### Dealing with Null values 

```python
# Check the data type to make correct decision. 
df[['Team','Contract']].dtypes
```

```python
# how many null values there are. 
df[['Team','Contract']].isnull().sum()
```

```python
# Since we don't know the value we will replace it with unknow. 
df['Team'].fillna("unknown", inplace= True )
df['Contract'].fillna("unknown", inplace= True)
```

```python
# check to see if there are Null Values.
df['Team'].isna().sum(), df['Contract'].isna().sum()
```
