```python
# Let's Replace the NaN Values with 0 since 
df["Hits"].fillna(0, inplace=True)
```

```python
df[df["Hits"]==0].head(10)
```

```python
# I want to change the data type to be INT. Let's check to see if we can.
df["Hits"].unique()
```

```python
# there are a few values that looke like 1.1K and we need to deal with the if we are to proceed.
# I wrote a python function to convert the numbers. 
def convert_to_int(value):
    if isinstance(value, str) and 'K' in value:
        return int(float(value.replace('K', '')) * 1000)
    elif isinstance(value, (int, float)):
        return int(value)
    else:
        return value
```

```python
# apply the function to the column
df["Hits"] = df["Hits"].apply(convert_to_int)
```

```python
# Some still makybe strings, so let's make sure they are not.
df["Hits"] = df["Hits"].astype(int)
```

## Height and Weight

#### Height

##### Missing Values check


```python
df["Height"].isna().sum()
```

```python
df["Height"].head()
```

```python
#now we can split it and multiply for the value. 
df["Height"].head()
```

```python
# we must write a function to convert the number to CM
def convert_to_cm(height):
    feet , inches = height.split("'")
    cm = (int(feet) * 12 + int(inches[:-1])) * 2.54
    return round(cm, 2) 
```

```python
# Apply the function to convert the Height column
df['Height_cm'] = df["Height"].apply(convert_to_cm)
df['Height_cm']
```

```python
# Let's Drop the original Height column as we now don't need it
df.drop(columns = ["Height"], inplace = True)
```

### Weight

Similarly we can going to convert the weight to metric as well.
right now it's in pounds (Imperial) and we want it in Kilos (Metric).

```python
df["Weight"].head()
```

```python
# Let's check for missing Values
df["Weight"].isna().sum()
```

```python
# Now we write a similar python script to convert the weight
def convert_to_kg(weight):
    pounds = int(weight[:-3])
    kg = pounds * 0.453592
    return round(kg, 2)
```

```python
df["Weight_kg"] = df["Weight"].apply(convert_to_kg)
```

```python
df["Weight_kg"].head()
```

```python
# now we can drop the old weight colmn
df.drop(columns = ["Weight"], inplace = True)
```
